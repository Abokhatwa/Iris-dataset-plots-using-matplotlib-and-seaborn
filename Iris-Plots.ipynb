```python
from sklearn import datasets
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
iris = datasets.load_iris()
df = pd.DataFrame(iris.data, columns = ['Sepal Length','Sepal Width','Petal Length', 'Petal Width'])
df['Label']=iris.target
```

```python
x = df[df['Label']==0]
plt.figure(figsize=(7,7))
plt.hist(x['Sepal Length'], color="red")
plt.hist(x['Sepal Width'], color="green")
plt.hist(x['Petal Length'], color="blue")
plt.hist(x['Petal Width'], color="yellow")

plt.legend(['Sepal Length','Sepal Width','Petal Length','Petal Width'])
plt.show()
```

```python
import seaborn as sns
corr = df.drop(['Label'], axis=1).corr()
plt.figure(figsize=(10,10))

sns.heatmap(corr,xticklabels=corr.columns.values,yticklabels=corr.columns.values,cmap='YlGnBu',annot = True)

```

```python
mapping ={0.0:'Iris-setosa',1.0:'Iris-versicolor',2.0:'Iris-virginica'} 
list_mapped=[mapping[i] for i in df.Label]
df['Species']=list_mapped
```
